Question 1: Explain the followings:
1.  Define the context window in the context of language models.
2.  Explain why the model's context window is typically kept small.
3.  Identify and elaborate on which OpenAI and Gemini language model boasts the largest
context window and provide a detailed explanation.

1. A context window is the amount of text a language model can remember at once while generating a response. A larger context window allows the model to consider more previous information, making its answers more accurate and relevant.

2. The model's context window is kept small because a larger window increases processing time, memory usage, and computational cost. It also adds complexity to the hardware and training process, which can slow down the model and reduce its efficiency. A smaller context window helps the model work faster and more accurately.

3. OpenAI's GPT-4 Turbo supports a context window of **128,000 tokens**, while Gemini 1.5 Ultra boasts a window of **1 million tokens**. A larger context window lets the model handle longer text and maintain more context, though it requires more computational resources.

````python
def compare_text(text1, text2):

    if not text1.strip() or not text2.strip():
        return "Error: test1 oe tes2 is empty!"

    model = genai.GenerativeModel('gemini-1.5-pro')
    prompt = f"""
    Compare the following two texts and provide a similarity score (0% to 100%):
    1. {text1}
    2. {text2}

    Only return a numeric value. Do NOT return any explanations or code.
    """
    try:
        response = model.generate_content(prompt)
        similarity_score = response.text.strip().replace("```python", "").replace("```", "")
        return f"Texts ka Similarity Score: {similarity_score}%"
    except Exception as e:
        return f"API Request Failed: {str(e)}"
````

````python
def generate_blog_content(topic):

    prompts = {
        "title": f"Generate an engaging blog post title for the topic: {topic}",
        "description": f"Write a detailed, informative blog post on the topic '{topic}' in around 300 words.",
        "keywords": f"Provide a list of 5-7 SEO-related keywords for a blog post about {topic}.",
        "seo_meta_title": f"Generate an SEO-friendly meta title (under 60 characters) for a blog about {topic}.",
        "seo_meta_description": f"Write a short SEO meta description (under 160 characters) for a blog about {topic}."
    }

    blog_data = {}

    for key, prompt in prompts.items():
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[{"role": "system", "content": "You are a helpful blog content generator."},
                      {"role": "user", "content": prompt}]
        )

        blog_data[key] = response["choices"][0]["message"]["content"].strip()

    return blog_data
````

````python
def safe_generate_blog(topic):
    if not topic.strip():
        return "Error: Topic cannot be empty. Please enter a valid blog topic."

    try:
        blog_content = generate_blog_content(topic)
        return blog_content
    except Exception as e:
        print(f"Error generating blog content: {e}")
        return "API Request Failed: Please check your API key and network connection.", ""
````

````python

def display_blog(blog_content):

    if isinstance(blog_content, dict):
        print("\n Blog Title:")
        print(blog_content["title"])

        print("\n Blog Description:")
        print(blog_content["description"])

        print("\n Related Keywords:")
        print(blog_content["keywords"])

        print("\n SEO Meta Title:")
        print(blog_content["seo_meta_title"])

        print("\n SEO Meta Description:")
        print(blog_content["seo_meta_description"])
    else:
        print(blog_content)
````

````python
topic= input("enter a blog data ")
blog_content = safe_generate_blog(topic)
display_blog(blog_content)
````

names and locations of these companies from the text.



````python
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", grouped_entities=True)
````

````python
article=input("enter a article: ")
entities = ner_pipeline(article)
companies = [entity['word'] for entity in entities if entity['entity_group'] == 'ORG']
locations = [entity['word'] for entity in entities if entity['entity_group'] == 'LOC']
print("\nInformation:")
print("Companies:", companies)
print("Locations:", locations)
````
